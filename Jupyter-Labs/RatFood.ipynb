```python
def getMaze(n):
    maze = []
    for _ in range(n):
        maze.append([False]*n)
    return maze
```

```python
maze = getMaze(4)
```

 You are given a maze once you have visited that place you must not visit that place again.

```python
# def mazePath(ratpostion, foodposition, maze, Solution)
def mazePath(ratC, ratR, foodR, foodC, maze,solution=""):
    # base case
    # it should not allow the rat to go out of maze
    if ratR < 0 or ratC < 0 or ratR >= len(maze) or ratC >= len(maze[0]):
        return
    
    # should not go already visited place
    if maze[ratR][ratC]:
        return
    
    # cord of rat is same as cord of food
    if (ratR, ratC) == (foodR, foodC):
        print(solution)
        return
    
    maze[ratR][ratC] = True
#     solution.append("N")
    mazePath(ratR - 1, ratC, foodR, foodC, maze,solution+"N")
    mazePath(ratR, ratC - 1  , foodR, foodC, maze,solution+"E")
    mazePath(ratR + 1, ratC, foodR, foodC, maze,solution+"S") 
    mazePath(ratR , ratC + 1, foodR, foodC, maze,solution+"W")
    maze[ratR][ratC] = False
```

```python
mazePath(0, 0, 3, 3, maze)
```

```python
# def mazePath(ratpostion, foodposition, maze, Solution)
def mazePathPrint(ratC, ratR, foodR, foodC, maze,solution=[]):
    # base case
    # it should not allow the rat to go out of maze
    if ratR < 0 or ratC < 0 or ratR >= len(maze) or ratC >= len(maze[0]):
        return
    
    # should not go already visited place
    if maze[ratR][ratC]:
        return
    
    # cord of rat is same as cord of food
    if (ratR, ratC) == (foodR, foodC):
        print(solution)
        return
    
    maze[ratR][ratC] = True
    solution.append("N")
    mazePathPrint(ratR - 1, ratC, foodR, foodC, maze,solution)
    solution.pop()
    
    solution.append("E")
    mazePathPrint(ratR, ratC - 1  , foodR, foodC, maze,solution)
    solution.pop()
    
    solution.append("S")
    mazePathPrint(ratR + 1, ratC, foodR, foodC, maze,solution) 
    solution.pop()
    
    solution.append("W")
    mazePathPrint(ratR , ratC + 1, foodR, foodC, maze,solution)
    solution.pop()
    
    maze[ratR][ratC] = False
```

```python
mazePathPrint(0,0,3,3, maze)
```
